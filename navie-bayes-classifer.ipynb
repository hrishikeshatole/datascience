```python
data = pd.read_csv('../input/pimaindian/pima-indians-diabetes.csv')
```

```python
data.head()
```

```python
y = data['class']
x = data.drop(['class'], axis= 1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30, random_state=1)
```

```python
nb_model = GaussianNB()
nb_model.fit(x_train, y_train)
```

```python
nb_predict = nb_model.predict(x_test)
```

```python
print('Accuracy of gaussian NB', nb_model.score(x_test, y_test))
```
