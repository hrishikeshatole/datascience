```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import svm
```

```python
data = pd.read_csv('../input/letterdata/letterdata.csv')
data.head()
```

```python
def getaccuracy(testset, pred):
    correct = 0
    for x in range(len(testset)):
        if testset[x] == pred[x]:
            correct+=1
    return (correct/float(len(testset))*100.0)
```

```python
#splitting data
x,y = np.array(data)[:, 1:16], np.array(data.letter)[:]
```

```python
x_train = x[: 16000, :] #the , is seperation for row and column
x_test = x[16001:, :]
y_train = y[:16000]
y_test = y[16001:]
```

```python
# building model
model = svm.SVC(gamma=0.025, C=3) # generally used 0.025 and c 3 
#but its not an hard and fast rule
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print("accuracy", getaccuracy(y_test,y_pred))
```
